```python
#global path1
def video_to_frame(path1): # Enter path where the video file is present
    global vid 
    vid=cv2.VideoCapture(r'{}'.format(path1)) 
    # frame 
    currentframe = 0
    while(True):
        # reading from frame 
        ret,frame = vid.read()
        #cv2.waitKey(100)
    
        if ret:
            name = './files/frame{}'.format(currentframe)+ '.jpg'
            print ('Creating-->' + name) 
  
            # writing the images extracted from video 
            cv2.imwrite(name, frame) 
            currentframe += 1
        else:
            break
vid.release()
cv2.destroyAllWindows()
```

```python
def img_preprocess(path): #path='C:\\Users\\Prashant\\Desktop\\tensorflow notebooks\\task\\files'
    # grab the paths to the input images and initialize our images list
    print("loading images...")
    global images 
    images = []
    for im in glob(path+'\\frame*.jpg'):
        image = cv2.imread(im)
    images.append(image)
    return images

#initialize OpenCV's image sticher object and then perform the image
print("Stitching images object created...")
stitcher =cv2.Stitcher.create()
```
